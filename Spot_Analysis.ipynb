```python
playlist = p.drop(columns=['Artist IDs','Valence','Instrumentalness','Speechiness','Time Signature','Key','Tempo','Added By','Spotify ID','Artist IDs'])
```

```python
playlist.info()
```

```python
playlist
```

```python
playlist['Added At']=pd.to_datetime(playlist['Added At']) 
playlist['Added At'] = playlist['Added At'].dt.strftime('%Y-%m-%d')
```

```python
playlist['Release Date']=pd.to_datetime(playlist['Release Date']) 
playlist['Release Date'] = playlist['Release Date'].dt.strftime('%Y-%m-%d')
```

```python
playlist['Danceability'] = playlist['Danceability'] * 100
```

```python
playlist.head()
```

```python
release = playlist.sort_values(by = 'Release Date',ascending=False)
```

```python
release.head()
```

```python
Artists = playlist.groupby('Artist Name(s)')['Track Name'].count()
Artists
```

```python
top_artists = Artists.sort_values(ascending=False)
```

```python
top_artists = top_artists.head(20)
top_artists
```

```python
plt.figure(figsize=(10,5))
top_artists.plot(kind='bar',color='skyblue')
plt.xlabel('Artist Name')
plt.ylabel('No. of Tracks')
plt.title('Top 20 Artists with Maximum No. of Tracks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
